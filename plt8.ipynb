fifa data

```python
fifa = pd.read_csv('./fifa_data.csv')
fifa.head(7)
```

Histogram

```python
bins = [40,50,60,70,80,90,100] # x axis pr show hoga
plt.hist(fifa.Overall,bins, color='grey') # sirf 40,60 ... ayese print krega 
plt.xticks(bins) # print all bins 40,50,60.....

plt.ylabel('Number of player')
plt.xlabel('Skill Level')
plt.title('Distribution of player skills in FIFA 2018')
plt.show()
```

Pie chart

```python
plt.title('Foot Preference of FIFA Players')
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']
plt.pie([left,right], labels=labels , colors=colors, autopct='%.2f %%') # %% lagane se 23.19% show hoga

plt.show()
```

Explode

```python
y = np.array([35,25,15,12])
mylabels = ['apple', 'chiku','orange','dates']
myexplode= [0.2, 0, 0, 0]

plt.pie(y, labels= mylabels, explode=myexplode)
plt.show()
```

pie chart- 2

```python
# fifa.Weight # show data in lbs
fifa.Weight = [int (x.strip('lbs'))if type(x)==str else x for x in fifa.Weight] 
# fifa.Weight
# fifa.Weight[0]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 225) & (fifa.Weight <150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight <175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

# plt.style.use('ggplot')
weights = [light , light_medium, medium, medium_heavy, heavy]
labels= ['Under 125', '125-150', '150-175', '175-200', 'over 200']
myexplode = {0.2, 0, 0, 0, 0} # piece me kr dega  unde 125 and 125-150 ko
plt.pie(weights, labels=labels,explode =myexplode, autopct='%.2f %%', pctdistance= 0.8)
plt.legend()

plt.show()
```
