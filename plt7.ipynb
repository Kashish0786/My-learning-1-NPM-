```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Basic graph

```python
x = [1,2,3]
y= [2,4,6]
# resize graph, dpi = pixel
plt.figure(figsize=(5,3), dpi=100)
# plt.plot(x,y, label='2x', color = 'red' , linewidth= 2 ,linestyle='--' ,marker='.' , markersize = 12,
        #  markeredgecolor ='blue')
        
# use shorthand notation = ['color'][marker][line]

plt.plot(x,y, 'r.--' , label= '2x')

#line nu. 2
# x2= np.arange(0,4,0.5)
# print(x2)
# plt.plot(x2, x2**2, 'r', label='X^2')

# line nu. 2
x2= np.arange(0,4,0.5)
# print(x2)
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')
plt.plot(x2[5:], x2[5:]**2, 'b--')

plt.title("first graph" , fontdict={'fontname' : 'comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,6.5,7,10])

plt.legend() # label dikhane ke liye use this 
plt.savefig("mygraph.png", dpi=100) #file me save ho jaayega ye graph 
plt.show()
```

Bar chart

```python
labels = ['a', 'b','c']
values =[1,4,2]

bars =plt.bar(labels,values)

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

# insted of above we can also use 
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(5,3))
```
