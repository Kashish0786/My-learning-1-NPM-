List vs numpy array

```python
py_list = [1,2,3]
print("python list multiplication", py_list * 2)

np_array = np.array([1,2,3])
print("python array multiplication", np_array * 2)

import time 
start = time.time()
py_list = [i*2 for i in range (1000000)]
print ("\n list operation time : ", time.time() - start)

start = time.time()
np_array = np.arange(1000000)*2
print("\n numpy operation time : ", time.time()- start)
```

CREATING ARRAY FROM SCRATCH 

```python
zero = np.zeros((3,4))
print("zero array: \n ", zero )

ones = np.ones((2,5))
print("ones array: \n ", ones )

full = np.full((9,4), 7)
print("full array: \n ", full )
# random array important 
random = np.random.random((4,5))
print("random array: \n ", random )
# sequence arrray important 
sequence = np.arange(0,10,2)
print("sequence array: \n ", sequence )
```

VECTOR, MATRIX, TENSOR


```python
vector = np.array([1,2,3])
print("Vector: \n ", vector )

matrix = np.array,([[1,2,3],
                    [4,5,6]])
print("matrix: \n ", matrix )

tensor = np.array([[[1,2], [3,4]], 
                   [[5,6], [7,8]] ])
print("tensor: \n ", tensor )
```

Array Properties

```python
arr = np.array([[1,2,3],[4,5,6]] )
print("shape: \n ", arr.shape )
print("dimension: \n ", arr.ndim)
print("size: \n ", arr.size)
print("Dtype: \n ", arr.dtype )
```

ARRAY RESHAPING 

```python
arr = np.arange(12)
print("orignal array", arr)

reshaped = arr.reshape((3,4))
print("\n reshape", reshaped)
# it returns copy
flattened = reshaped.flatten()
print("\n flattened array", flattened)

#  revel (return view , insted of copy)
raveled = reshaped.ravel()
print("\n reveled array", raveled)

# transpose
transpose = reshaped.T
print("\n transpose array", transpose)

```
