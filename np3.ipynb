```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data structure [resturant_id,2021,2022]
sales_data = np.array( [
    [1, 19000, 22000], #x hotel
    [2, 11000, 12000], #y hotel
    [3, 13000, 25000], #s hotel
    [4, 15000, 24000], #g hotel
    [5, 18000, 25000] #k hotel
   
])

print("Zomato sales analysis")
print("\n Sales data shape", sales_data.shape) #give shape of matrix 
print("\n sample data for 1st 3 hotel", sales_data[:3]) # 3 row and colum print krwayega
```

```python
# total sales per year
print(np.sum(sales_data, axis=0))
year_total = np.sum(sales_data[:,1:], axis= 0) # [:,1:] saari rows aur 1st cloum add nhi hoga
print(year_total)

#min sales per resturent , row ke liye axis = 1, column ke liye axis = 0
min_sales = np.min(sales_data[:, 1:], axis = 1)
print(min_sales)

# max sale per year
max_sales = np.max(sales_data[:, 1:], axis = 0)
print(max_sales)

# avg sales per resturant 
avg_sales = np.mean(sales_data[:, 1:], axis = 0)
print(avg_sales)

#cumalitive sums 
cumsum = np.cumsum(sales_data[:, 1:], axis = 1)
print(cumsum)

plt.figure(figsize=(6, 3)) # 6,3 map kis size ka hoga
plt.plot(np.mean(cumsum, axis=0))
plt.title("Avg cumulitive sales across all resturent ")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3])
vector2 = np.array([4,5,6])

print("vector addition", vector1 + vector2)
print("vector multiply", vector1 * vector2)
print("dot product", np.dot(vector1, vector2))

# np.linalg.norm(vector1) it will give magnitude of vec1, formula= sin(a,b) = a.b / | |a|.|b| |
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```

```python
# vectorized operation , individually sabke saath jaakr operation perform karega 
resturant_types = np.array(['biryani', 'chineese', 'pizza', 'cafe'])
vectorized_upper = np.vectorize(str.upper)
print("vectorized upper", vectorized_upper(resturant_types))
```
